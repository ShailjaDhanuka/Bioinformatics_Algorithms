```python
import sys

# Please do not remove package declarations because these are used by the autograder.

MASSES = {
    "G": 57, "A": 71, "S": 87, "P": 97, "V": 99, "T": 101, "C": 103, "I": 113, "L": 113,
    "N": 114, "D": 115, "K": 128, "Q": 128, "E": 129, "M": 131, "H": 137, "F": 147,
    "R": 156, "Y": 163, "W": 186,
}

def get_combination(peptide,length):
    patterns=[]
    for i in range(len(peptide)):
        pattern=peptide[i:i+length]
        if len(pattern)<length:
           circularise = length-len(pattern)
           pattern+=peptide[:circularise]
        patterns.append(pattern)
    #print(patterns)
    return patterns
              
# Insert your cyclopeptide_scoring function here, along with any subroutines you need
def cyclopeptide_scoring(peptide: str, spectrum: list[int]) -> int:
    i=len(peptide)-1
    all_combinations=[]
    while i>0:
        all_combinations+=get_combination(peptide,i)
        #print(i)
        i-=1
    #print(all_combinations)
    all_masses=[] 

    for combination in all_combinations:
        mass=0
        for aa in combination:
            mass+=MASSES[aa]
        all_masses.append(mass)
    mass=0        
    for aa in peptide: # for the full mass
        mass+=MASSES[aa]
    all_masses.append(mass)
        
    all_masses.append(0)
    all_masses.sort()
    

    score=0
    for num in spectrum:
        if num in all_masses:
            score+=1
            all_masses.remove(num)

    return score  
```
