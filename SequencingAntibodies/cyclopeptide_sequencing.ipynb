```python
import sys
MASSES = {
    "G": 57, "A": 71, "S": 87, "P": 97, "V": 99, "T": 101, "C": 103, "I": 113, "L": 113,
    "N": 114, "D": 115, "K": 128, "Q": 128, "E": 129, "M": 131, "H": 137, "F": 147,
    "R": 156, "Y": 163, "W": 186,
}

alphabet=""
for key in MASSES:
    alphabet+=key
# print(alphabet)

peptide="NQEL"
def LinearSpectrum(peptide,alphabet,MASSES):
    preMass=[0]
    for i in range(len(peptide)):
        for letter in alphabet:
            
            if letter==peptide[i]:
                preMass.append(preMass[i]+ MASSES[letter])
                
    # print(preMass)
    linear_spectrum=[0]
    for i in range(0,len(peptide)):
        for j in range(i+1,len(peptide)+1):
            # print(i,j,preMass[j],preMass[i])
            linear_spectrum.append(preMass[j]-preMass[i])
    return sorted(linear_spectrum)
def CyclicSpectrum(peptide,alphabet,MASSES):
    preMass=[0]
    for i in range(len(peptide)):
        for letter in alphabet:
            
            if letter==peptide[i]:
                
                preMass.append(preMass[i]+ MASSES[letter])
    peptideMass=preMass[len(peptide)]
                
    # print(preMass)
    cyclic_spectrum=[0]
    for i in range(0,len(peptide)):
        for j in range(i+1,len(peptide)+1):
            # print(i,j,preMass[j],preMass[i])
            cyclic_spectrum.append(preMass[j]-preMass[i])
            if i>0 and j<len(peptide):
                cyclic_spectrum.append(peptideMass-(preMass[j]-preMass[i]))
    return sorted(cyclic_spectrum)
def Expand(candidate_peptides,new_candidates):
    
    newcandidates=[]
    # print(candidate_peptides,new_candidates)
    for aa1 in new_candidates:
        
        for aa2 in candidate_peptides:
            # print(aa1,aa2)
            newcandidates.append(aa1+aa2)
            
    return newcandidates

def FinalPeptides(Spectrum):
    peptides=set()
    for i in range(len(Spectrum)-1):
        for j in range(i+1, len(Spectrum)):
            value = Spectrum[j] - Spectrum[i]
            if 50 < value < 200:
                for k, v in MASSES.items():
                    if v == value:
                        peptides.add(k)
    return peptides

def isconsistent(spectrum,pep_spec):
    # if spectrum[-1]<pep_spec[-1]:
    #     return False
    spectrum_copy = spectrum[:]   # make a copy

    for mass in pep_spec:
        if mass in spectrum_copy:
            spectrum_copy.remove(mass)
        else:
            return False

    return True

def PepStr(pep):
    pepstr=[]
    for aa in pep:
        pepstr.append(MASSES[aa])
    return pepstr

# Insert your cyclopeptide_sequencing function here, along with any subroutines you need
def cyclopeptide_sequencing(spectrum: list[int]) -> list[list[int]]:
    result=[]
    candidate_peptides=FinalPeptides(spectrum)
    # print(candidate_peptides)
    new_candidates=candidate_peptides.copy()
    # print(new_candidates,candidate_peptides)
    
    
    while len(new_candidates)!=0:
        new_candidates=Expand(candidate_peptides,new_candidates)
        # print(new_candidates)
        # print(new_candidates)
       
        for pep in new_candidates[:]:
            pep_spec = LinearSpectrum(pep,MASSES=MASSES,alphabet=alphabet)
            pep_str=PepStr(pep)
            pep_cyc_spec = CyclicSpectrum(pep,MASSES=MASSES,alphabet=alphabet)
            
            if pep_spec[-1]==spectrum[-1]:
                # print(pep_cyc_spec)
                if pep_cyc_spec==spectrum and pep_str not in result:
                    result.append(pep_str)
                    # print("pep kept", pep)

                new_candidates.remove(pep)
                # print(new_candidates)
            elif not isconsistent(spectrum,pep_spec):
                # print("pep removed", pep)
                new_candidates.remove(pep)
            
        #print(len(new_candidates), "is the length of new candidates",new_candidates)
    return result



```
