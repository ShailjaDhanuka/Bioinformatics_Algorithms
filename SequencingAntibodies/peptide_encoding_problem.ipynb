```python
import sys

# Please do not remove package declarations because these are used by the autograder.

GENETIC_CODE = {
    "AAA": "K", "AAC": "N", "AAG": "K", "AAU": "N", "ACA": "T", "ACC": "T", "ACG": "T",
    "ACU": "T", "AGA": "R", "AGC": "S", "AGG": "R", "AGU": "S", "AUA": "I", "AUC": "I",
    "AUG": "M", "AUU": "I", "CAA": "Q", "CAC": "H", "CAG": "Q", "CAU": "H", "CCA": "P",
    "CCC": "P", "CCG": "P", "CCU": "P", "CGA": "R", "CGC": "R", "CGG": "R", "CGU": "R",
    "CUA": "L", "CUC": "L", "CUG": "L", "CUU": "L", "GAA": "E", "GAC": "D", "GAG": "E",
    "GAU": "D", "GCA": "A", "GCC": "A", "GCG": "A", "GCU": "A", "GGA": "G", "GGC": "G",
    "GGG": "G", "GGU": "G", "GUA": "V", "GUC": "V", "GUG": "V", "GUU": "V", "UAA": "",
    "UAC": "Y", "UAG": "", "UAU": "Y", "UCA": "S", "UCC": "S", "UCG": "S", "UCU": "S",
    "UGA": "", "UGC": "C", "UGG": "W", "UGU": "C", "UUA": "L", "UUC": "F", "UUG": "L",
    "UUU": "F",
}

def find_patterns(dna,peptide):
    all_matches=[]
    for i in range(0,len(dna) - len(peptide)*3 + 1):
        pattern=dna[i:i+len(peptide)*3]
        # print(i,pattern)
        kmer_peptide=""
        for j in range(0,len(pattern),3):
            codon=pattern[j:j+3]
            aminoacid=GENETIC_CODE[codon]
            kmer_peptide+=aminoacid
        if kmer_peptide==peptide:
            all_matches.append(pattern)
    return all_matches

# Insert your peptide_encoding function here, along with any subroutines you need
def peptide_encoding(text: str, peptide: str) -> list[str]:
    sequence=text
    sequence=sequence.replace("T","U")
    reverse_sequence=sequence[-1:len(sequence)*-1-1:-1]
    reverse_sequence = reverse_sequence.translate(str.maketrans("AUGC", "UACG"))
    
    result=[]
    for_result = find_patterns(sequence,peptide)
    for pattern in for_result:
        pattern=pattern.translate(str.maketrans("U", "T"))
        result.append(pattern)
    # print("rev now")
    rev_result = find_patterns(reverse_sequence,peptide)
     
    for pattern in rev_result:
        # print(pattern)
        pattern=pattern[::-1]
        # print(pattern)
        pattern=pattern.translate(str.maketrans("AUGC", "TACG"))
        result.append(pattern)
    
    
    return result
```
