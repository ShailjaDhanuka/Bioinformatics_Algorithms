```python
import sys
from typing import List, Dict, Iterable
import random
import copy

# Please do not remove package declarations because these are used by the autograder.
def Cycle(g,start):
    
    path = [start]
    while start in g:
        
        next_node = g[start].pop()
        path.append(next_node)

        if g[start]==[]:
            g.pop(start)
        start = next_node

    return path,g



# Insert your eulerian_cycle function here, along with any subroutines you need
# g[u] is the list of neighbors of the vertex u
def eulerian_cycle(g: Dict[int, List[int]]) -> Iterable[int]:
    """Constructs an Eulerian cycle in a graph."""
    g = {u: g[u][:] for u in g} 
    
    possible_starts = list(g)
    start = random.choice(possible_starts)
    possible_starts.remove(start) #randomize a start location
    #first cycle
    cycle,remaining = Cycle(g,start)
    #print(cycle,remaining)

    while remaining: # there are points left in the remaining dictionary
        for i, v in enumerate(cycle):
            if v in remaining:
                new_cycle, remaining = Cycle(remaining, v)
                # print(new_cycle)
        
        #now add the splices
                cycle = cycle[:i] + new_cycle + cycle[i+1:]
                break
        

    return cycle

        
```
