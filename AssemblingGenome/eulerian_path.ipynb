```python
import sys
from typing import List, Dict, Iterable

# Please do not remove package declarations because these are used by the autograder.

# Insert your eulerian_path function here, along with any subroutines you need
# g[u] is the list of neighbors of the vertex u
import sys
from typing import List, Dict, Iterable
import random
import copy

# Please do not remove package declarations because these are used by the autograder.
def Path(g,start):
    
    path = [start]
    #print(g,start)
    while start in g:
        
        next_node = g[start].pop()
        path.append(next_node)

        if g[start]==[]:
            g.pop(start)
        start = next_node

    return path

# 0: 2
# 1: 3
# 2: 1
# 3: 0 4
# 6: 3 7
# 7: 8
# 8: 9
# 9: 6

def starting_node(g):
    in_deg={}
    out_deg={}
    for u in g:
        out_deg[u] = len(g[u])
        in_deg.setdefault(u, 0)
        for v in g[u]:
            in_deg[v] = in_deg.get(v, 0) + 1

    start = None
    #mke cycle
    for u in in_deg | out_deg:
        o = out_deg.get(u, 0)
        i = in_deg.get(u, 0)
        if i == o + 1: # this node is the last node
            end_node = u
            #print(end_node)
        if o==i+1: # this node is the first node
            start_node = u
            #print(start_node)
    # g.setdefault(end_node, []).append(start_node)
    return start_node,end_node
    


# Insert your eulerian_cycle function here, along with any subroutines you need
# g[u] is the list of neighbors of the vertex u
def eulerian_path(g: Dict[int, List[int]]) -> Iterable[int]:
    """Constructs an Eulerian cycle in a graph."""
    #copy g
    g = {u: g[u][:] for u in g} 
    start, end = starting_node(g)

    #add a fake edge bw start and end
    g.setdefault(end,[]).append(start)

    cycle = Path(g,start)


    while g: # there are points left in the remaining dictionary

        for i, v in enumerate(cycle):
            if v in g:
                # print(remaining)
                new_cycle=Path(g,v)
                cycle  = cycle[:i] + new_cycle + cycle[i+1:]

                break

    return cycle[:-1]

        
```
