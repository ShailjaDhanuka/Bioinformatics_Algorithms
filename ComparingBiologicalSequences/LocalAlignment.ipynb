```python
import sys
from typing import List, Dict, Iterable, Tuple

import sys
from typing import List, Dict, Iterable, Tuple

# Please do not remove package declarations because these are used by the autograder.

# Insert your GlobalAlignment function here, along with any subroutines you need
import sys
import numpy as np

import sys
import numpy as np

def LCSBackTrack(match_reward,mismatch_penalty,indel_penalty,v,w):
  s = np.zeros((len(v)+1, len(w)+1), dtype=int)

#   for i in range(1, len(v)+1):
#         s[i][0] = s[i-1][0] - indel_penalty

#   for j in range(1, len(w)+1):
#         s[0][j] = s[0][j-1] - indel_penalty
# #   print(s)

  Backtrack = np.zeros((len(v)+1,len(w)+1),dtype=object)

  for i in range(1,len(v)+1):
     for j in range(1,len(w)+1):
         match = 0
         if v[i-1] == w[j-1]:
             match = match_reward
         else:
             
             match = -mismatch_penalty
         s[i][j] = max(0,s[i-1][j]-indel_penalty, s[i][j-1]-indel_penalty , s[i-1][j-1] + match )
         #print(s[i][j])
         if s[i][j] == s[i-1][j]-indel_penalty:
             Backtrack[i][j] = "↓"
         elif s[i][j] == s[i][j-1]-indel_penalty:
             Backtrack[i][j] = "→"
         elif s[i][j]== s[i-1][j-1] + match:
             Backtrack[i][j] = "↘"
  i, j = np.unravel_index(np.argmax(s), s.shape)
  #print(s)
  return Backtrack, s ,i,j

 
def OutputAlignment(backtrack, v, w, i, j,s):
    #print(len(backtrack),len(backtrack[1]))
    #print(i,j)
    if s[i][j] == 0:
        return "", ""
    #insert the case that if it starts with a 0 then it is the beginning (or something like that, even in the arrows, gotta do something about that)

    if backtrack[i][j] == "↘":
        av, aw = OutputAlignment(backtrack, v, w, i-1, j-1,s)
        return av + v[i-1], aw + w[j-1]
    elif backtrack[i][j] == "→":
        av, aw = OutputAlignment(backtrack, v, w, i, j-1,s)
        return av + "-", aw + w[j-1]

    elif backtrack[i][j] == "↓":
        av, aw = OutputAlignment(backtrack, v, w, i-1, j,s)
        return av + v[i-1], aw + "-"

    
    
def LocalAlignment(match_reward: int, mismatch_penalty: int, indel_penalty: int,
                    s: str, t: str) :
    v=s
    w=t
    a,b,i,j = LCSBackTrack(match_reward,mismatch_penalty,indel_penalty,v,w)
 
    c,d = OutputAlignment(a,v,w,i,j,b)

    # str1,str2 = OutputAlignment(backtrack,v,w,i,j)
    # print(i,j)
    # print(scoreing_matrix)
    return (int(b[i][j]),c,d)
```
