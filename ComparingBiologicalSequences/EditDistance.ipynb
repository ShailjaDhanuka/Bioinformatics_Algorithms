```python
import sys
from typing import List, Dict, Iterable, Tuple

# Please do not remove package declarations because these are used by the autograder.

# Insert your EditDistance function here, along with any subroutines you need
def EditDistance(v: str, w: str) -> int:
    import numpy as np
    
    s = np.zeros((len(v)+1, len(w)+1), dtype=int)

    for i in range(len(v)+1):
        s[i][0] = i
    for j in range(len(w)+1):
        s[0][j] = j

    for i in range(1, len(v)+1):
        for j in range(1, len(w)+1):
            penalty = 0 if v[i-1] == w[j-1] else 1

            s[i][j] = min(
                s[i-1][j] + 1,        # delete
                s[i][j-1] + 1,        # insert
                s[i-1][j-1] + penalty    # substitute
            )

    return int(s[len(v)][len(w)])
```
