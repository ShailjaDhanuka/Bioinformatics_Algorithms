```python
import sys
from typing import List, Dict, Iterable, Tuple
import numpy as np

# Please do not remove package declarations because these are used by the autograder.

def LCSBackTrack(match_reward,mismatch_penalty, indel_penalty, v,w):
  s = np.zeros((len(v)+1,len(w)+1))
  Backtrack = np.empty((len(v)+1,len(w)+1),dtype=object)
  for j in range(1, len(w)+1):
    s[0][j] = s[0][j-1] - indel_penalty
  for i in range(1,len(v)+1):
     for j in range(1,len(w)+1):
            
         match=0
         if v[i-1] == w[j-1]:
             match = match_reward
         else:
             match = -mismatch_penalty
         
         s[i][j] = max(s[i-1][j]-indel_penalty, s[i][j-1]-indel_penalty , s[i-1][j-1] + match )
         #print(s[i][j])
         if s[i][j] == s[i-1][j]-indel_penalty:
             Backtrack[i][j] = "↓"
         elif s[i][j] == s[i][j-1]-indel_penalty:
             Backtrack[i][j] = "→"
         elif s[i][j]== s[i-1][j-1] + match:
             Backtrack[i][j] = "↘"

  i = len(v)
  j = np.argmax(s[i,:])
  max_score = s[i][j]
  return Backtrack , s,i,j,max_score


def OutputLocal(backtrack, v, w, i, j):
    if j == 0 or i==0:
        return "", ""

    if backtrack[i][j] == "↓":
        top, bottom = OutputLocal(backtrack, v, w, i-1, j)
        return top + v[i-1], bottom + "-"

    elif backtrack[i][j] == "→":
        top, bottom = OutputLocal(backtrack, v, w, i, j-1)
        return top + "-", bottom + w[j-1]

    else:  # diagonal
        top, bottom = OutputLocal(backtrack, v, w, i-1, j-1)
        return top + v[i-1], bottom + w[j-1]

# Insert your FittingAlignment function here, along with any subroutines you need
# Note that the function receives table BLOSUM as an argument:


def OverlapAlignment(match_reward: int, mismatch_penalty: int, indel_penalty: int,
                    s: str, t: str) -> Tuple[int, str, str]:
    
    backtrack, score_matrix, i, j, max_score = LCSBackTrack(match_reward,mismatch_penalty, indel_penalty, s, t)
    s1,t1 = OutputLocal(backtrack,s,t,i,j)
    return int(max_score),s1,t1
```
